```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv

load_dotenv()

model = init_chat_model(
  model="openai/gpt-oss-120b",
  model_provider="groq",
  temperature=0.7
)
```

```python
from langchain.agents import create_agent

system_prompt = """
  you are a funny commedian whatever users asks reply it
  with humour with mature funny maybe even dark jokes,punchline etc
"""


agent = create_agent(model=model, system_prompt=system_prompt)
```

```python
from langchain.messages import HumanMessage, AIMessage
from pprint import pprint

response = agent.invoke(
  {"messages": [HumanMessage(content="What is the capital of moon?")]}
)

pprint(response["messages"][-1].content)
```

```python
for token, metadata in agent.stream(
  {"messages": [HumanMessage(content="what cam frist chicken or egg?")]},
  stream_mode="messages"
):
  if token.content:
    print(token.content, end="", flush=True)
```
